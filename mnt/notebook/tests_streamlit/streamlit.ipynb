```python
from pyngrok import ngrok
import time
import os
import threading
from dotenv import load_dotenv

load_dotenv()
NGROK_TOKEN = os.getenv("NGROK_TOKEN")
ngrok.set_auth_token(NGROK_TOKEN)

public_url = ngrok.connect(8501)
print(f"Acesse o app Streamlit em: {public_url}")

def start_streamlit():
    os.system("streamlit run /content/drive/MyDrive/github_projects/openweather-pyspark-nosql/mnt/notebook/tests_streamlit/app.py")

thread = threading.Thread(target=start_streamlit)
thread.start()
```

```python
from pymongo import MongoClient
import pandas as pd
import datetime
import os
from dotenv import load_dotenv

load_dotenv()
# Conexão com MongoDB
MONGO_URI = os.getenv("MONGO_URI")

client = MongoClient(MONGO_URI)
db = client["clima"]
collection = db["dados"]
dados = list(collection.find())
df = pd.DataFrame(dados)

# Conversão de datas
df["data_coleta"] = pd.to_datetime(df["data_coleta"])
df["data_hora"] = pd.to_datetime(df["data_hora"])
```
